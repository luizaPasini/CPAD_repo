```python
##mostra se existem dados faltantes ou não
dados_faltantes = dados.isnull().values.any()
print (dados_faltantes)
if (dados_faltantes == True):
  print ("Existem dados faltantes")
else:
  print ("Não existem dados faltantes")
```

```python
##dados faltantes por coluna
faltantes = dados.isnull().sum()
faltantes = faltantes[faltantes > 0]
print("Colunas com dados faltantes:")
print(faltantes)
print(f"\nTotal de valores faltantes no DataFrame: {faltantes.sum()}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=dados,
    x="Island",
    y="Culmen Length (mm)",
    hue="Species"
)


plt.title("Culmen Length por Ilha e Espécie de Pinguim")
plt.ylabel("Culmen Length (mm)")
plt.xlabel("")
plt.legend(title="Species", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=dados,
    x="Culmen Length (mm)",
    y="Culmen Depth (mm)",
    hue="Species",
    palette="deep"
)


plt.title("Comprimento vs Profundidade do Culmen por Espécie de Pinguim")
plt.xlabel("Culmen Length (mm)")
plt.ylabel("Culmen Depth (mm)")


plt.legend(title="Species", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
